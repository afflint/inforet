```python
import pandas as pd
import pymongo
from datetime import datetime
import numpy as np
```

```python
def to_float(x):
    try:
        y = float(x)
    except ValueError:
        y = np.nan
    return y
```

```python
stations = 'data/weatherww2/stations.csv'
weather = 'data/weatherww2/weather.csv'
```

```python
S = pd.read_csv(stations, low_memory=False)
W = pd.read_csv(weather, converters={
    'Date': lambda x: datetime.strptime(x, '%Y-%m-%d'),
    'Snowfall': lambda x: to_float(x)
}, low_memory=False)
```
